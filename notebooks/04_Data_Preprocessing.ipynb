```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.compose import ColumnTransformer

from sklearn.pipeline import Pipeline

from sklearn.preprocessing import (
    OneHotEncoder,
    StandardScaler
)

import joblib
```

```python
df = pd.read_csv(
    "../data/processed/vessel_arrival_delay_feature_engineered.csv"
)

df.head()
```

```python
feature_columns = [

    "Carrier",
    "Vessel_Type",
    "Origin_Port",
    "Destination_Port",

    "Distance_NM",
    "Scheduled_Transit_Days",
    "Departure_Delay_Hours",
    "Average_Speed_Knots",
    "Vessel_Age",
    "Cargo_Load_Percentage",
    "Port_Congestion_Index",
    "Weather_Severity",
    "Fuel_Price_USD",
    "Historical_Route_Delay",

    "Season",
    "Customs_Clearance_Risk",

    "Route",
    "Congestion_Level",
    "Weather_Category",
    "Vessel_Age_Group",
    "Load_Category",
    "Speed_Category",

    "Delay_Risk_Score"
]
```

```python
target_column = "Late_Arrival"
```

```python
X = df[feature_columns]

y = df[target_column]
```

```python
X_train, X_test, y_train, y_test = train_test_split(

    X,
    y,

    test_size=0.20,

    random_state=42,

    stratify=y
)
```

```python
numerical_features = [

    "Distance_NM",
    "Scheduled_Transit_Days",
    "Departure_Delay_Hours",
    "Average_Speed_Knots",
    "Vessel_Age",
    "Cargo_Load_Percentage",
    "Port_Congestion_Index",
    "Weather_Severity",
    "Fuel_Price_USD",
    "Historical_Route_Delay",
    "Delay_Risk_Score"
]
```

```python
categorical_features = [

    "Carrier",
    "Vessel_Type",
    "Origin_Port",
    "Destination_Port",

    "Season",
    "Customs_Clearance_Risk",

    "Route",
    "Congestion_Level",
    "Weather_Category",
    "Vessel_Age_Group",
    "Load_Category",
    "Speed_Category"
]
```

```python
preprocessor = ColumnTransformer(

    transformers=[

        (

            "num",

            StandardScaler(),

            numerical_features

        ),

        (

            "cat",

            OneHotEncoder(

                handle_unknown="ignore"

            ),

            categorical_features

        )

    ]
)
```

```python
X_train_processed = preprocessor.fit_transform(
    X_train
)

X_test_processed = preprocessor.transform(
    X_test
)
```

```python
print(X_train_processed.shape)

print(X_test_processed.shape)
```

```python
joblib.dump(

    preprocessor,

    "../models/preprocessor.pkl"

)
```

```python
joblib.dump(
    X_train,
    "../models/X_train.pkl"
)

joblib.dump(
    X_test,
    "../models/X_test.pkl"
)

joblib.dump(
    y_train,
    "../models/y_train.pkl"
)

joblib.dump(
    y_test,
    "../models/y_test.pkl"
)
```
